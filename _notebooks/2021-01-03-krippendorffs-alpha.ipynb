## Inter-Annotator Agreement

The wisdom of crowds usually results in answers more accurate than any one individual. So a high-level of overall dataset agreement or Krippendorff alpha score is an indicator that the dataset is reliable for model training.
Krippendorff's alpha is a metric used in the social sciences for measure level of agreement in surveys or census data. The metric ranges from [-1, 1] and a score greater than 0.8 indicates high agreement. 

```python
import numpy as np
import pandas as pd

np.random.seed(42)
```

Create a random dataframe of annotations. Each row is a task and each column is an annotator. There are five label types ranging from 1 through 4. A 0 indicates that the specific task was not labeled by the annotator.

```python
annotators = ["a", "b", "c", "d", "e"]
tasks = [f"Task {i}" for i in np.arange(11) + 1]

# create a random dataframe of annotations 
annotations_mat = np.random.randint(0,5,size=(11,5))
annotations_df = pd.DataFrame(columns=annotators, index=tasks, data=annotations_mat)
annotations_df
```

### Expected Agreement

```python
counts_df = annotations_df.apply(pd.Series.value_counts, axis=1).drop(columns=[0]).fillna(0)
counts_per_label = counts_df.sum(axis=0)
counts_per_label
```

```python
total_annotations =  counts_df.sum().sum()
```

```python
label_relative_frequency = counts_per_label / counts_per_label.sum()
label_relative_frequency
```

```python
expected_agreement = np.sum(label_relative_frequency ** 2)
expected_agreement
```

### Actual Agreement

```python
counts_df = annotations_df.apply(pd.Series.value_counts, axis=1).drop(columns=[0]).fillna(0)
counts_df
```

```python
avg_annotations_per_task = total_annotations / len(tasks)
```

```python
counts_per_task = counts_df.sum(axis=1)
```

```python
sum_pairwise_agreements = (counts_df * (counts_df - 1)).sum(axis=1)
```

```python
average_agreement = np.mean(sum_pairwise_agreements / (avg_annotations_per_task * (counts_per_task - 1)))
```

```python
average_agreement
```

```python
k_alpha = (average_agreement - expected_agreement) / (1 - expected_agreement)
```
